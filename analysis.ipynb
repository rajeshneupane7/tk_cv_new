```python
main_path="/home/rajesh/work/tk_cv/data"

temp_store=[]
for file_name in os.listdir(main_path):
    if file_name.startswith('Da') & file_name.endswith('.csv'):
        df= pd.read_csv(os.path.join(main_path, file_name))
        df.columns= df.iloc[0]
        df= df[1:]
        df= df.reset_index(drop=True)
        temp_store.append(df)
```

```python
final_df= pd.concat(temp_store)
```

```python
final_df['date_time']= pd.to_datetime(final_df['Date Time'])
```
