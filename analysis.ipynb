```python
import pandas as pd

# Example
cow_df['date_time'] = pd.to_datetime(
    cow_df['Year'].astype(str) + '-' +
    cow_df['Month'].astype(str).str.zfill(2) + '-' +
    cow_df['Day'].astype(str).str.zfill(2) + ' ' +
    cow_df['Time'].astype(str)
)

```

```python
tempr_df['date_time']=pd.to_datetime(tempr_df['date_time_cst_cdt'])
```

```python
cow_df['pen']= cow_df['Channel'].apply(lambda x: 'Pen 2' if x.endswith('CH1') else
('Pen 3' if x.endswith('CH3')else
('Pen 4' if x.endswith('CH5') else 
'Pen 5'
)))
```

```python
cow_df.date_time.isna().sum()
```

```python
# 1. Standardize Timezones (Remove timezone info from both to make them match)
# We convert to datetime first (just in case), then strip the timezone.
cow_df['date_time'] = pd.to_datetime(cow_df['date_time']).dt.tz_localize(None)
tempr_df['date_time'] = pd.to_datetime(tempr_df['date_time']).dt.tz_localize(None)

# 2. SORTING IS MANDATORY for merge_asof
cow_df = cow_df.sort_values('date_time')
tempr_df = tempr_df.sort_values('date_time')

# 3. The Merge
final_df = pd.merge_asof(
    left=cow_df,
    right=tempr_df,
    left_on='date_time',     # Time column (must be sorted)
    right_on='date_time',
    left_by='pen',           # Grouping ID
    right_by='pen_id',       # Grouping ID
    direction='nearest',     # Finds the closest time match
     # Optional: Prevents matching if times are too far apart
)

print("Merge successful.")
```

```python
# Convert 'temperature_c' and 'rh' columns to numeric type
final_df["temperature_c"] = pd.to_numeric(final_df["temperature_c"], errors="coerce")
final_df["rh"] = pd.to_numeric(final_df["rh"], errors="coerce")

# Now perform the THI calculation
final_df["THI"] = (0.8 * final_df["temperature_c"]) + ((final_df["rh"] / 100) * (final_df["temperature_c"] - 14.4)) + 46.4
```

```python
final_df.plot(y='total_cows_area', x='THI', kind='scatter')
```
